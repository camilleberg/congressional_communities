```python

def render_and_open(chart, filename: str = "chart.html"): 
    """ 
    Renders a Pyecharts chart to ./output/filename and opens it in the de 
    
    Args: 
        chart (Chart): A Pyecharts chart instance. 
        filename (str): Output HTML filename (default: "chart.html"). 
        
    Source: https://medium.com/@shouke.wei/visualizing-data-with-pyecharts-a-complete-tutorial-with-real-world-examples-421b548e9633
    """ 
    
    output_dir = "figures" 
    
    os.makedirs(output_dir, exist_ok=True) # Ensure output directory exiats
    output_path = os.path.join(output_dir, filename) 
    
    chart.render(output_path) 
    file_path = os.path.abspath(output_path) 
    webbrowser.open(f"file://{file_path}")
```

+ this is an example

```python


columns = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
data1 = [2.6, 5.9, 9.0, 26.4, 28.7, 70.7, 175.6, 182.2, 48.7, 18.8, 6.0, 2.3]  # precipitation
data2 = [2.0, 4.9, 7.0, 23.2, 25.6, 76.7, 135.6, 162.2, 32.6, 20.0, 6.4, 3.3]  # evaporation

bar = (
    Bar()
    .add_xaxis(columns)
    .add_yaxis("precipitation", data1)
    .add_yaxis("evaporation", data2)
    .set_global_opts(title_opts=opts.TitleOpts(title="Bar chart"))
)

bar = (
    Bar()
    .add_xaxis(columns)
    .add_yaxis("precipitation", data1)
    .add_yaxis("evaporation", data2)
    .set_global_opts(title_opts=opts.TitleOpts(title="Bar chart"))
)

line = (
    Line()
    .add_xaxis(columns)
    .add_yaxis("precipitation", data1, label_opts=opts.LabelOpts(is_show=True))
    .add_yaxis("evaporation", data2, label_opts=opts.LabelOpts(is_show=True))
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title="Line chart",
            subtitle="Annual precipitation and evaporation",
            pos_top="45%",
        )
    )
)

grid = (
    Grid()
    .add(bar, grid_opts=opts.GridOpts(pos_bottom="60%"))
    .add(line, grid_opts=opts.GridOpts(pos_top="60%"))
)

render_and_open(grid, "grid.html")
```

```python
# 1. Load your custom GeoJSON
with open(geo_json_path, "r", encoding="utf-8") as f:
    custom_geojson = json.load(f)

# 2. Your data — names must match the "name" property inside your GeoJSON's "features"
# region_data = [
#     ("Region A", 100),
#     ("Region B", 80),
#     ("Region C", 60),
# ]
# using the population data from the GeoDataFrame
pop_data = dict(zip(gdf.GEOID, gdf.D001TPOP))

# 3. Build the map, registering the geojson via a JS function call
custom_map = (
    Map()
    .add_js_funcs(
        "echarts.registerMap('my_custom_map', {});".format(json.dumps(custom_geojson))
    )
    .add(
        series_name="my data",
        data_pair=list(pop_data.items()),
        maptype="my_custom_map",   # must match the name used in registerMap above
        is_map_symbol_show=False,
    )
    
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Custom Map Example"),
        visualmap_opts=opts.VisualMapOpts(max_=100),
    )
)

render_and_open(custom_map, "grid.html")
```

```python
from ipyleaflet import Map, GeoJSON
from ipywidgets import Output, HBox, Layout
from IPython.display import display, HTML
from pyecharts.charts import Bar
from pyecharts import options as opts

# Example data — swap this dict for your real cenpy/Census pull, keyed by GEOID
region_data = {
    "06037": {"name": "Los Angeles County, CA", "years": [2019, 2020, 2021, 2022],
              "population": [10039107, 10014009, 9829544, 9721138]},
    "48201": {"name": "Harris County, TX", "years": [2019, 2020, 2021, 2022],
              "population": [4713325, 4731145, 4728030, 4780913]},
    "17031": {"name": "Cook County, IL", "years": [2019, 2020, 2021, 2022],
              "population": [5150233, 5085639, 5028092, 5023480]},
}

# Minimal example GeoJSON — replace with your real file via gpd.read_file(...).__geo_interface__
example_geo = {
    "type": "FeatureCollection",
    "features": [
        {"type": "Feature", "properties": {"GEOID": geoid, "NAME": d["name"]},
         "geometry": {"type": "Point", "coordinates": [lon, lat]}}
        for geoid, d, (lon, lat) in zip(
            region_data.keys(), region_data.values(),
            [(-118.24, 34.05), (-95.37, 29.76), (-87.63, 41.88)]
        )
    ],
}

geojson_layer = GeoJSON(data=example_geo, point_style={"radius": 8, "fillColor": "#378ADD", "fillOpacity": 0.8})
out = Output(layout=Layout(width="500px", height="400px"))

def handle_click(event=None, feature=None, **kwargs):
    with out:
        out.clear_output(wait=True)
        geoid = feature["properties"]["GEOID"]
        d = region_data[geoid]
        chart = (
            Bar()
            .add_xaxis([str(y) for y in d["years"]])
            .add_yaxis("Population", d["population"])
            .set_global_opts(title_opts=opts.TitleOpts(title=f"Population — {d['name']}"))
        )
        display(HTML(chart.render_embed()))

geojson_layer.on_click(handle_click)

m = Map(center=(38, -97), zoom=4, layout=Layout(width="600px", height="500px"))
m.add_layer(geojson_layer)

display(HBox([m, out]))
```
